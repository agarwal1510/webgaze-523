```python
import pandas as pd
import matplotlib
import warnings
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import mlab
import math
import seaborn as sns
```

```python
df_4g = pd.DataFrame(columns=['sitename','category','median_scroll', 'size', 'onload_uplt', 'implicit_uplt', 'median_uplt'])
df_scroll = pd.read_csv('max_pct_viewports_scrolled_multi_4g.csv')
df_size = pd.read_csv('mobile_screens_multi_4g.csv')
df_onload = pd.read_csv('mobile_scroll_screen_viewports_onload_4g_max.csv')
df_implicit = pd.read_csv('scroll_implicit_uplt_multi_4g_median.csv')
df_median = pd.read_csv('scroll_uplt_multi_4g_median.csv')


df_4g['sitename'] = df_scroll['sitename']

# df_3g = df_3g.sort_values('sitename')
# df_onload = df_onload.sort_values('sitename')
# df_implicit = df_implicit.sort_values('sitename')
# df_median = df_median.sort_values('sitename')
# df_size = df_size.sort_values('site')
# df_size = df_size.reset_index(drop=True)

df_4g['onload_uplt'] = df_onload['value']
df_4g['implicit_uplt'] = df_implicit['value']
df_4g['median_uplt'] = df_median['value']
df_4g['size'] = df_size['numScreens']
df_4g['category'] = df_onload['category']

df_scroll['mean'] = df_scroll.mean(axis=1)
df_4g['median_scroll'] = df_scroll['mean']

df_4g
```

```python
df_cor = {'Scroll':df_4g['median_scroll'], 'site_size':df_4g['size'], 
          'onLoad_uplt':df_4g['onload_uplt'], 'Implicit_uplt':df_4g['implicit_uplt'], 
          'Median_uplt':df_4g['median_uplt']}
df_cor = pd.DataFrame(df_cor).astype('float64')
r = df_cor.corr()
sns.heatmap(r, fmt=".2g", annot=True, cmap="viridis", linewidths=.8)
plt.show()
```

```python
df_4g.plot(kind='bar', x ='sitename', y ='onload_uplt',title= 'OnLoad')
plt.rcParams["figure.figsize"] = (20,5)
axes = plt.gca()
ymin, ymax = axes.get_ylim()
plt.show()

df_4g.plot(kind='bar', x ='sitename', y ='implicit_uplt',title= 'Implicit')
plt.rcParams["figure.figsize"] = (20,5)
axis = plt.gca()
axis.set_ylim([0, ymax])
plt.show()

df_4g.plot(kind='bar', x ='sitename', y ='median_uplt',title= 'Median')
plt.rcParams["figure.figsize"] = (20,5)
axiss = plt.gca()
axiss.set_ylim([0, ymax])
plt.show()
```
